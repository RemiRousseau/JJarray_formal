```python
import quick_eq_imp
import analysis

import numpy as np
import matplotlib.pyplot as plt
```

```python
res = []
N_l = [10,20,30]
for n in N_l:
    y_eqs = quick_eq_imp.equivalent_impedance(n,all_eqs = False,impedance = False,mass_capa = False)
    res.append(analysis.circuit_resonances(y_eqs[-1],full = False))
```

```python
fig,ax = plt.subplots(2,1,figsize = (12,10))
n_round = 6
remove_doublons = [list(set([(round(el[0],n_round),round(el[1],n_round)) for el in itera])) for itera in res]
for el in remove_doublons:
    el.sort()
for i,n in enumerate(N_l):
    len_f = len(remove_doublons[i])
    #xs = [(val)/len_f for val in range(len_f)]
    xs = list(range(len_f))
    ax[0].plot(xs,[el[0] for el in remove_doublons[i]])
    ax[1].plot(xs,[el[1] for el in remove_doublons[i]])
```
