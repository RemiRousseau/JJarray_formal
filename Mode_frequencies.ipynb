```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 20
w0 = 15*10**9
l_j = 15*10**(-9)/N
c_j = 1/(l_j*w0**2)
c_0 = 10**-12
```

```python
freqs = []

for c_0,N in zip([10**-11,10**-11*(5/8)],[80,50]):
    f = []
    for k in range((N+1)//2):
        f.append(w0*np.sqrt((1-np.cos(np.pi*k/N))/(c_0/(2*N*c_j)+1-np.cos(np.pi*k/N))))
    freqs.append(f)
for el in freqs:
    plt.plot(el)
```

```python
N_s = [10*(i+1) for i in range(4)]
```

```python
freqs = []
for N in N_s:
    f = []
    c_0 = 10**-11/N
    for k in range((N+1)//2):
        f.append(w0*np.sqrt((1-np.cos(np.pi*k/N))/(c_0/(2*N*c_j)+1-np.cos(np.pi*k/N))))
    freqs.append(f)
```

```python
for j,N in enumerate(N_s):
    plt.plot(freqs[j])
plt.xlim(0,5)
```

```python
plt.plot([0,0.5],[w0,w0],"-r", label = "w0")
for N in np.linspace(20,1000,1):
#for N in np.logspace(1,3,20):
    freqs = []
    N = int(N)
    for k in range((N+1)//2):
        freqs.append(w0*np.sqrt((1-np.cos(np.pi*k/N))/(c_0/(2*N*c_j)+1-np.cos(np.pi*k/N))))
    plt.plot([i/N for i in range((N+1)//2)],freqs,"-+",label = str(N))
plt.legend(bbox_to_anchor =(1.0, 1.25))
plt.xlabel('k/N')
plt.ylabel('Mode Frequency')
plt.show()
```
